# Limpieza de datos

En esta primera parte del proyecto del curso de machine learning se realiza la limpieza de la base de datos. Primero se explora el conjunto de datos.  

```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")
datos.head(10)
```

```python
tipos = datos.dtypes
print(tipos)
```

```python
import pandas as pd

# Mapear las categorías a unos (1) y ceros (0)
datos['family_history_with_overweight'] = datos['family_history_with_overweight'].map({'yes': 1, 'no': 0})
datos['FAVC'] = datos['FAVC'].map({'yes': 1, 'no': 0})
datos['Gender'] = datos['Gender'].map({'Female': 1, 'Male': 0})
datos['SMOKE'] = datos['SMOKE'].map({'yes': 1, 'no': 0})
datos['SCC'] = datos['SCC'].map({'yes': 1, 'no': 0})
datos['CAEC'] = datos['CAEC'].map({'no': 0, 'Sometimes': 1, 'Frequently':2,'Always':3})
datos['CALC'] = datos['CALC'].map({'no': 0, 'Sometimes': 1, 'Frequently':2,'Always':3})
datos['MTRANS'] = datos['MTRANS'].map({'Public_Transportation': 0, 'Automobile': 1, 'Walking':2,'Bike':3, 'Motorbike':4})
datos['Obesity'] = datos['NObeyesdad'] 
datos['Obesity'] = datos['Obesity'].map({'Normal_Weight': 0, 'Insufficient_Weight': 0, 'Obesity_Type_I':1,'Obesity_Type_II':1, 'Obesity_Type_III':1, 'Overweight_Level_I':0,'Overweight_Level_II':0})
#datos['NObeyesdad'] = datos['NObeyesdad'].map({'Normal_Weight': 0, 'Insufficient_Weight': 1, 'Obesity_Type_I':2,'Obesity_Type_II':3, 'Obesity_Type_III':4, 'Overweight_Level_I':5,'Overweight_Level_II':6})
datos['']

datos.head(10)
```

```python
pd.crosstab(datos.NObeyesdad, datos.Obesity)
frecuencias = datos['Age'].value_counts().reset_index()
frecuencias.columns = ['Age', 'Frecuencia']
print(frecuencias)
```

```python
datos['Age'] = round(datos['Age'] )
datos['FCVC'] = round(datos['FCVC'] )
datos['NCP'] = round(datos['NCP'] )
datos['CH2O'] = round(datos['CH2O'] )
datos['FAF'] = round(datos['FAF'] )
datos['TUE'] = round(datos['TUE'] )

datos.head(10)
```
