```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colormaps
from sklearn import datasets
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
obesidad =pd.read_csv('obesidad.csv')
obesidad.head()
```

```python
### Gráficos de dispersión
plt.rcParams["figure.figsize"] = (13,6) 

# grafica 1
plt.subplot(2,3,1)
plt.scatter(obesidad['Height'], obesidad['Weight'], c=obesidad['Obesity'])
plt.savefig('scatter.png')
plt.xlabel('Estatura')
plt.ylabel('Peso')
plt.grid()

# grafica 2
plt.subplot(2,3,2)
plt.scatter(obesidad['Age'], obesidad['Weight'], c=obesidad['Obesity'])
plt.savefig('scatter.png')
plt.xlabel('Edad')
plt.ylabel('Peso')
plt.grid()

# grafica 3
plt.subplot(2,3,3)
plt.scatter(obesidad['CH2O'], obesidad['Weight'], c=obesidad['Obesity'])
plt.savefig('scatter.png')
plt.xlabel('Consumo de agua')
plt.ylabel('Peso')
plt.grid()

# grafica 4
plt.subplot(2,3,4)
plt.scatter(obesidad['NCP'], obesidad['Weight'], c=obesidad['Obesity'])
plt.savefig('scatter.png')
plt.xlabel('Comidas diarias')
plt.ylabel('Peso')
plt.grid()

# grafica 5
plt.subplot(2,3,5)
plt.scatter(obesidad['MTRANS'], obesidad['Weight'], c=obesidad['Obesity'])
plt.savefig('scatter.png')
plt.xlabel('Medio de transporte')
plt.ylabel('Peso')
plt.grid()

# grafica 6
plt.subplot(2,3,6)
plt.scatter(obesidad['TUE'], obesidad['Weight'], c=obesidad['Obesity'])
plt.savefig('scatter.png')
plt.xlabel('Tiempo de uso objetos tecnológicos')
plt.ylabel('Peso')
plt.grid()

```
